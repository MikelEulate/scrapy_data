```python
base_url = 'https://www.regalos.es/top-ventas?includes=events%2Cproducts%2Cpersonalized'

def get_recursively(url):

    r = requests.get(url)
    c = r.content

    soup = BeautifulSoup(c, "html.parser")
    
    
    
    container = soup.find("div", class_ = 'category-products').ul
    elements = container.find_all('li')
    #lista_regalos = []
    for ele in elements:
        #print(ele.find('a')['title'])
        reg = ele.find('a')['title'].replace('\n', '')
        reg_text =  ele.find('a')['href'].split('.es/')[-1].split('.html')[0]
        #print(reg_text)
        lista_regalos.append(reg)
        price = ele.find('span', class_ = 'price').text.replace('\n', '')
        print(price)
        prices.append(price)
        if reg_text != '':
            prices.append(price)
            lista_regalos.append(reg_text)
    #pages = soup.find('div', class_ = 'pages').find_all('a', class_= 'next i-next')[0]['href']
    #print(pages)
    #try:
    #    get_recursively(pages)
    #except Exception as e:
    #    print(e)
    
    return lista_regalos
    
    
```
