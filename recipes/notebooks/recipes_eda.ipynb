```python
def corregir_df(df):
    
    def corregir_puntuacion(x):
        punt_1 = float(str(x).split('/')[0])
        punt_2 = float(str(x).split('/')[1])
        x_n = punt_1 / punt_2
        
        return x_n*10
    
    def corregir_votos(x):
        
        return int(str(x))
    
    def corregir_tiempo(x):
        tiempo = str(x).split('minutos')[0]
        try:
            return int(tiempo)
        except:
            return 0
    
    def corregir_n_personas(x):
        personas = str(x).split('Para')[-1].split('personas')[0]
        
        try:
            return int(personas)
        except:
            return 0
    
    def corregir_precio_persona(x):
        personas = str(x).split(' €/persona')[0]
        personas = personas.replace(',', '.')
        try:
            return float(personas)
        except:
            return 0
    
    def corregir_kcalorias(x):
        try:
            kcalorias =float(x.split('kcal por 100g.')[0])
            return kcalorias
        except:
            return 0
    
    
    df['puntuacion'] = df['puntuacion'].apply(lambda x : corregir_puntuacion(x))
    df['votos'] = df['votos'].apply(lambda x : corregir_votos(x))
    df['tiempo'] = df['tiempo'].apply(lambda x : corregir_tiempo(x))
    df['n_personas'] = df['n_personas'].apply(lambda x : corregir_n_personas(x))
    df['precio_persona'] = df['precio_persona'].apply(lambda x : corregir_precio_persona(x))
    df['kcalorias'] = df['kcalorias'].apply(lambda x : corregir_kcalorias(x))
    
    return df
```

```python
df_new = corregir_df(df)
```
